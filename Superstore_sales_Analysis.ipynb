```python
import pandas as pd
df = pd.read_csv('Sample - Superstore.csv' , encoding='latin1')
df.head()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
import matplotlib.pyplot as plt
category_sales.plot(kind='bar', title='Total Sales by Category', color='orange')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

subcategory_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values()
subcategory_profit.plot(kind="barh", color='mediumseagreen', title="Profit by Sub-Category")
plt.xlabel("Total Profit")
plt.tight_layout()
plt.show()
```

```python
region_data = df.groupby("Region")[["Sales", "Profit"]].sum()
region_data.plot(kind="bar", figsize=(8,5), color=["skyblue", "salmon"], title="Region-wise Sales & Profit")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Discount"], df["Profit"], color='purple', alpha=0.6)
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True)
plt.tight_layout()
plt.show()
```
